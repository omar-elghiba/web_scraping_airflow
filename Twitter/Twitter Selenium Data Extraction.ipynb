```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from datetime import datetime, timedelta
import time
import re
```

```python
def last_week_date():
    
    lst = []
    today = datetime.now()
    
    now = today.strftime('%Y-%m-%d')

    b = today - timedelta(days=7)

    one_week_ago = b.strftime('%Y-%m-%d')
    
    lst.append(one_week_ago)
    lst.append(now)
    
    return lst
```

```python
def twitter_login(user_name,pass_word):
    
    driver = webdriver.Chrome('Path')

    driver.maximize_window()

    driver.get('https://twitter.com/i/flow/login')
    
    username = WebDriverWait(driver, 10).until(
    EC.visibility_of_element_located((By.TAG_NAME, 'input'))
    )
    username.send_keys(user_name)

    button1= driver.find_elements(By.XPATH, '//div[@role="button"]')[-2].click()

    password = WebDriverWait(driver, 10).until(
        EC.visibility_of_element_located((By.XPATH, "//input[@type='password']"))
    )
    password.send_keys(pass_word)
    
    button= driver.find_elements(By.XPATH, '//div[@role="button"]')[-1].click()
    
    return driver
```

```python
def search_accounts(account):
    
    driver = twitter_login(user_name,pass_word)
    
    all_accounts_data = {}
    

        
    search = WebDriverWait(driver, 10).until(
        EC.visibility_of_element_located((By.XPATH, "//input[@role='combobox']"))
    )

    week_date = last_week_date()

    search.send_keys(f'from:{account} since:{week_date[0]} until:{week_date[1]}')

    search.send_keys(Keys.RETURN)

    articles = WebDriverWait(driver, 30).until(
        EC.visibility_of_all_elements_located((By.TAG_NAME, 'article')))

    account_data_list = []
    
    
    account_data_list.append({'Followers' : 0})

    for article in articles:

        article_data = {}

        try:
            text_element = WebDriverWait(article, 10).until(EC.visibility_of_element_located((By.XPATH, ".//div[@data-testid='tweetText']")))

            text = text_element.text

            article_data['text'] = text  
        except:
            article_data['text'] = None 

        try:
            video = WebDriverWait(article, 10).until(EC.visibility_of_element_located((By.XPATH, ".//div[@data-testid='videoComponent']")))

            article_data['video'] = True  
        except:
            article_data['video'] = None 

        try:
            image = WebDriverWait(article, 10).until(EC.visibility_of_element_located((By.XPATH, ".//div[@data-testid='tweetPhoto']")))

            article_data['image'] = True  
        except:
            article_data['image'] = None 

        a = article.text.split('\n')


        a = article.text.split('\n')


        article_data['Hashtag'] = re.findall(r'#\w+', article_data['text'])

        article_data['View'] = a[-1] if len(a) > 0 else None

        article_data['Like'] = a[-2] if len(a) > 1 else None

        article_data['Retweet'] = a[-3] if len(a) > 2 else None

        article_data['Reply'] = a[-4] if len(a) > 3 else None

        account_data_list.append(article_data)

    all_accounts_data[account] = account_data_list
    
    driver.close()

    return all_accounts_data

```
